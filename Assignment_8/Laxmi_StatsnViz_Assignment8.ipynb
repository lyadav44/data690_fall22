Cleanse the data to make the year, population, cost, cost per person numerical type


--Code functions to cleanse the columns. One function for one column


--use apply() method of data frame to apply the functions to the columns for data cleansing

```python
# Cleaning Pupoluation

def clean_population(col_name):
  return int(col_name.replace(",",""))

df['Total Population'] = df['Total Population'].apply(clean_population)
df.head()
```

```python
# Cleaning Census Cost

def clean_census_cost(df):
    df['Census Cost'] = df['Census Cost'].map(lambda x: x.replace('$',''))
    df['Census Cost'] = df['Census Cost'].map(lambda x: x.replace(',',''))
    return df
df = clean_census_cost(df)

df.head()
```

```python
# Cleaning Average Cost Per Person
def clean_cost_per_person(df):
    df['Average Cost Per Person'] = df['Average Cost Per Person'].map(lambda x: x.replace('cents',''))
    df['Average Cost Per Person'] = df['Average Cost Per Person'].map(lambda x: x.replace('$',''))
    return df
df = clean_cost_per_person(df)

df.head()
```

```python
# Cleaning Year
def clean_year(df):
    df['Census Year'] = df['Census Year'].map(lambda x: x.replace('*',''))
    return df
df = clean_year(df)

df.head(23)
```

Use Plotly to plot the average cost per person over time.


```python
import pandas as pd

print("Pandas Version is ", pd.__version__)

import plotly 

print("Plotly Version is", plotly.__version__)

import plotly.graph_objects as go
import plotly.express as px
```

```python
fig = go.Figure()

trace_cost = go.Bar(      
    x=df["Census Year"],    
    y=df["Average Cost Per Person"],
    name="India"
)

fig.add_trace(trace_cost)


fig.update_layout(
    title="Average Cost Per Person Over Time",
    xaxis_title="Census Year",
    yaxis_title="Average Cost Per Person"
)

fig.show()
```

Use Plotly to plot the scatter plot of Total Population vs Total Cost.


```python
fig = px.scatter(
    df,
    x="Average Cost Per Person",
    y="Total Population", 
    title="Total Cost VS Total Populaton", 
    color="Average Cost Per Person", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```
