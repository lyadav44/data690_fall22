Cleanse the data to make the year, population, cost, cost per person numerical type


--Code functions to cleanse the columns. One function for one column


--use apply() method of data frame to apply the functions to the columns for data cleansing

```python
# Cleaning Pupoluation, cost, cost per person numerical type

def Census_year(a):
  a = a.replace('*', '')
  return int(a)

def Total_population(a):
  a = a.replace(',', '')
  return int(a)

def Census_cost(a):
  if 'Billion' in a:
    a = a.replace('$', '')
    a = a.replace(',', '')
    a = a.replace(' Billion', '')
    a = float(a)
    return a*(10**9)
  else:
    a = a.replace('$', '')
    a = a.replace(',', '')
    return float(a)

def Average_cost_per_person(a):
  if ' cents' in a:
    a = a.replace('$', '')
    a = a.replace(' cents', '')
    a = float(a)
    a = a/100
    return a
  else:
    a = a.replace('$', '')
    return float(a)
```

```python
df['Census Year Update'] = df['Census Year'].apply(Census_year)
df['Total Population Update'] = df['Total Population'].apply(Total_population)
df['Census Cost Update'] = df['Census Cost'].apply(Census_cost)
df['Average Cost Per Person Update'] = df['Average Cost Per Person'].apply(Average_cost_per_person)
```

```python
df.drop(['Census Year', 'Total Population', 'Census Cost', 'Average Cost Per Person'],axis = 1,  inplace = True)
```

Use Plotly to plot the average cost per person over time.


```python
import pandas as pd

print("Pandas Version is ", pd.__version__)

import plotly 

print("Plotly Version is", plotly.__version__)

import plotly.graph_objects as go
import plotly.express as px
```

```python
fig = go.Figure()

trace_cost = go.Bar(      
    x=df["Census Year Update"],    
    y=df["Average Cost Per Person Update"]
 
)

fig.add_trace(trace_cost)


fig.update_layout(
    title="Average Cost Per Person Over Time",
    xaxis_title="Census Year Update",
    yaxis_title="Average Cost Per Person Update"
)

fig.show()
```
