```python
import pandas as pd
import plotly.express as px

df= pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv")
```

Choosing Population as Indicator

```python
df=df[df['indicator']=='SP.POP.TOTL']
df
```

Get data For the Specified Country Code

```python
country_code= ['AUS', 'CAN', 'FRA']
new_country_df=df[df['Country Code'].isin(country_code)]
new_country_df
```

Filter data for the year 2021

```python
new_country_df21=new_country_df[new_country_df['Year']==2021]
new_country_df21
```

### Bar plot: Population vs Country

```python
fig = px.bar(
    new_country_df21, 
    x="Country Name", 
    y="value", 
    title="2021 Population of Australia, France, Canada", 
    color="Country Name", 
    template="plotly_dark"
)

fig.show()
```

### Pie Chart

```python
fig = px.pie(
    new_country_df21, 
    names="Country Name", 
    values="value", 
    title="2021 Population of Australia, France, Canada", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)
fig.show()
```

### Box Plot

```python
fig = px.box(new_country_df21, x="Country Name", y="value", title='2021 Population of Australia, France, Canada')
fig.show()
```

### Comparing all years using Line Plot

```python
fig = px.line(new_country_df, x="Year", y="value", color="Country Name", title='Australia, France, Canada')
fig.show()
```

Comparing Average population for Different regions

```python
df.Region.unique()
```

```python
new_country_df21 = new_country_df21[new_country_df21.Year == 2021].groupby('Region').value.sum()
new_country_df21
```

Bar plot

```python
fig = px.bar(new_country_df21,x = new_country_df21.index, y = 'value', title = '2021 total population of the regions', color = new_country_df21.index)
fig.update_layout(showlegend = False)
fig.show()
```

### Pie Plot

```python
fig = px.pie(new_country_df21, values= 'value', names=new_country_df21.index, title='Pie chart for the 3 regions 2021 population')
fig.update_traces(textinfo='value+label')
fig.show()
```

```python
df_regions = new_country_df.groupby(['Region','Year'])['value'].sum()
df_regions
```

```python
df_regions = df_regions.reset_index()
df_regions
```

### Change in population over the years for different regions

```python
fig = px.line(df_regions, x='Year', y = 'value', color='Region')
fig.show()
```
