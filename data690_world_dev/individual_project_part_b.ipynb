```python
#reading csv file using pandas
long_df=pd.read_csv('https://raw.githubusercontent.com/lyadav44/data690_fall22/main/data690_world_dev/wdx_long.csv')
long_df.head(10)
```

```python
#Cleaning the data-
#dropping null value
long_df.drop(columns=["Unnamed: 0"], inplace=True)
long_df
```

```python

df_sa_countries = long_df[long_df["Country Code"].isin(['BTN','AFG','PAK','IND','LKA','NPL','BGD'])]
df_sa_countries
```

```python
df_se_edu = df_sa_countries.query("indicator == 'SE.SEC.ENRL'")
df_se_edu
```

```python
df_pivot = df_se_edu.pivot(index =['Year','Country Code','Country Name','Region','Income Group'],columns = 'indicator', values ='value').reset_index()
df_pivot.reset_index().head(10)
```

# Line graph - Total number of people who enrolled for Secondary Education

```python
fig = px.line(
    df_pivot, 
    x="Year", 
    y="SE.SEC.ENRL",
    labels={"SE.SEC.ENRL":"Secondary education, pupils"},
    hover_name='Country Name',
    color='Country Name'
    )
fig.show()
```

```python
df_Labor_force = df_sa_countries.query("indicator == 'SL.TLF.ACTI.1524.ZS'")
df_Labor_force
```

```python
df_pivot_lf = df_Labor_force.pivot(index =['Year','Country Code','Country Name','Region','Income Group'],columns = 'indicator', values ='value').reset_index()
df_pivot_lf.reset_index().head(10)
```

```python
fig = px.line(
    df_pivot_lf, 
    x="Year", 
    y="SL.TLF.ACTI.1524.ZS",
    labels={"SL.TLF.ACTI.1524.ZS":"Labor force participation rate for ages 15-24, total (%)"},
    hover_name='Country Name',
    color='Country Name',
    height=600
    )
fig.show()
```

```python
df_pivot_2 = df_sa_countries.pivot(index =['Year','Country Code','Country Name','Region','Income Group'],columns = 'indicator', values ='value').reset_index()
df_pivot_2.reset_index().head(10)
```

```python
df_pivot_2
fig = px.scatter(df_pivot_2, 
                x="SL.TLF.ACTI.1524.ZS", 
                y="SE.SEC.ENRL",
                title="'Percentage of Youth in Labor Force' VS 'Total number of people enrolled in higher level education'",
                labels={"SL.TLF.ACTI.1524.ZS":"Labor force participation rate for ages 15-24, total (%) (modeled ILO estimate)","SE.SEC.ENRL":"Secondary education, pupils"},
                color="Country Name",
                hover_name='Year',
                height=700,
                size_max=7,
                size="SL.TLF.ACTI.1524.ZS")
fig.update_traces(marker_size=40)
fig.show()
```

```python
df_pivot_4 =  df_pivot_lf[df_pivot_lf['Year'] == 2018]
fig_1 = px.bar(df_pivot_4, 
               x='Country Name', 
               y='SL.TLF.ACTI.1524.ZS',
               title='Labor force participation rate for ages 15-24, total (%) - 2018',
               labels={"SL.TLF.ACTI.1524.ZS":"2018 Labor force participation rate for ages 15-24, total (%) (modeled ILO estimate)"},
               hover_data=['SL.TLF.ACTI.1524.ZS'], 
               color='Country Name',
               height=800).update_xaxes(categoryorder="total descending")
fig_1.show()
```

```python
df_pivot_2018 = df_pivot[df_pivot['Year'] == 2018]
fig = px.bar(df_pivot_2018, x='Country Name', y='SE.SEC.ENRL',title='Total number of people enrolled in secondary level education - 2018'
             ,hover_data=['SE.SEC.ENRL'], color='Country Name',
             labels={'SE.SEC.ENRL':'2018 Secondary education, pupils'}, height=600).update_xaxes(categoryorder="total descending")
fig.show()
```
