**Find out the definition of the above 9 variables and document the definitions in the Jupyter Notebook in a Markdown cell so that the readers of your notebook will know the meaning of these variables.**

- **UNITID** : UNITID is an integer which represent the unit id of the institute
- **INSTNM** : INSTNM has school name and full form is Institution name. It index is fulltext.
- **CITY**: It is the city name where the institute are located. It is varchar accepting upto 200 char.
- **STABBR**: STABBR is the state postcode of the city where the school is located. It is accepts varchar upto 50 char.
- **ZIP**: Zip is the zipcode of the area where the institute resides. It accepts the varchar upto 20 char.
- **ADM_RATE**: ADM_RATE is the admission rate. It accepts the float value.
- **UGDS**: It represents the enrollment of undergraduate certificate/degree-seeking students. It accepts the float value.
- **TUITIONFEE_IN**: TUITIONFEE_IN is the in state tuition fee that the student are supposed to pay for the degree they are seeking. It takes an integer value
- **MN_EARN_WNE_P10**: It is the Mean earnings of students working and not enrolled 10 years after entry.
