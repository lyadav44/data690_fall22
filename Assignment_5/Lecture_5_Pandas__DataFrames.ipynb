## Modifying DataFrames



### Adding a new column

```python
langs = pd.Series(
    ['French', 'German', 'Italian'],
    index=['France', 'Germany', 'Italy'],
    name='Language'
)
```

```python
langs
```

```python
df['Language'] = langs
```

```python
df
```

---
### Replacing values per column

```python
df['Language'] = 'English'
```

```python
df
```

---
### Renaming Columns


```python
df.rename(
    columns={
        'HDI': 'Human Development Index',
        'Anual Popcorn Consumption': 'APC'
    }, index={
        'United States': 'USA',
        'United Kingdom': 'UK',
        'Argentina': 'AR'
    })
```

```python
df.rename(index=str.upper)
```

```python
df.rename(index=lambda x: x.lower())
```

---
### Dropping columns

```python
df.drop(columns='Language', inplace=True)
```

---
### Adding values

```python
df.append(pd.Series({
    'Population': 3,
    'GDP': 5
}, name='China'))
```

Appending returns a new `DataFrame`

```python
df
```

Setting the new index and values to the `DataFrame`:

```python
df.loc['China'] = pd.Series({'Population': 1_400_000_000, 'Continent': 'Asia'})
```

```python
df
```

Using `drop` to just remove a row by index:

```python
df.drop('China', inplace=True)
```

```python
df
```

---
### More radical index changes

```python
df.reset_index()
```

```python
df.set_index('Population')
```

## Creating columns from other columns



```python
df[['Population', 'GDP']]
```

The regular pandas way of expressing that, is just dividing each series:

```python
df['GDP'] / df['Population']
```

The result of that operation is just another series that you can add to the original `DataFrame`:

```python
df['GDP Per Capita'] = df['GDP'] / df['Population']
```

```python
df
```
